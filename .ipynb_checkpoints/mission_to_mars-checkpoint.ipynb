```python
from bs4 import BeautifulSoup as bs
import requests
import splinter
from splinter import Browser
import pandas as pd
```

```python
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://mars.nasa.gov/api/v1/news_items/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
resp = requests.get(url)
resp
```

```python
soup = bs(resp.text, 'html.parser')
soup
```

```python
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
```

```python
# target = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
# iresp = requests.get(target)
# image = bs(iresp.text, 'lxml')
html = browser.html
image = bs(html, 'lxml')
```

```python
image.find("section", {"class": "centered_text clearfix main_feature primary_media_feature single"})
```

```python
first = image.find("section", {"class": "centered_text clearfix main_feature primary_media_feature single"})
first
```

```python
jpg = first.a['data-fancybox-href']
jpg
```

```python
image_url = 'https://www.jpl.nasa.gov' + jpg
image_url
```

```python
browser.visit(image_url)
```

```python
facts_url = "https://space-facts.com/mars/"
facts_resp = requests.get(facts_url)
facts_soup = bs(facts_resp.text, 'html.parser')
```

```python
facts_table = facts_soup.find('table', {"class": "tablepress tablepress-id-comp-mars blue-table"})
```

```python
str(facts_table)
```
